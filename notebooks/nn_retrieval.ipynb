```python
""" Functions to work with the word embeddings
"""

import pickle
import numpy as np


def sentence_embedding():

        with open('../data/interim/proc_b_src_emb.p', 'rb') as fp:
                src_emb = pickle.load(fp)
        with open('../data/interim/proc_b_src_word.p', 'rb') as fp:
                embedding_dictionary_source = pickle.load(fp)

        with open('../data/interim/proc_b_trg_emb.p', 'rb') as fp:
                trg_emb = pickle.load(fp)
        with open('../data/interim/proc_b_trg_word.p', 'rb') as fp:
                embedding_dictionary_target = pickle.load(fp)

        def normalize_matrix(matrix):
                norms = np.sqrt(np.sum(np.square(matrix), axis=1))
                norms[norms == 0] = 1
                norms = norms.reshape(-1, 1)
                matrix /= norms[:]
                return matrix

        embedding_array_normalized_source = normalize_matrix(src_emb)
        embedding_array_normalized_target = normalize_matrix(trg_emb)

        return embedding_dictionary_source, embedding_array_normalized_source, embedding_dictionary_target, embedding_array_normalized_target
```

```python
embedding_dictionary_source, embedding_array_normalized_source, embedding_dictionary_target, embedding_array_normalized_target = sentence_embedding() 
```

```python
def calculate_nn(given_source_word, embedding_dictionary_source, embedding_array_normalized_source,
                 embedding_dictionary_target, embedding_array_normalized_target, n_neighbors=5):
                """
                Need normed embeddings!
                """
                given_source_index = embedding_dictionary_source[given_source_word]

                # Calculate Cos Similarity
                norm_src_word_emb = embedding_array_normalized_source[given_source_index]
                similarity_cos = np.dot(norm_src_word_emb, np.transpose(embedding_array_normalized_target))

                # Find Closest Neighbors
                most_similar_trg_index = np.argsort(-similarity_cos[[0]])[:n_neighbors].tolist()[0][:n_neighbors]
                
                inverse_trg_word = {index: word for word, index in embedding_dictionary_target.items()}
                
                target_word_list = []
                for single_neighbor in most_similar_trg_index:
                    target_word_list.append(inverse_trg_word[single_neighbor])
                    print(inverse_trg_word[single_neighbor])
                return target_word_list
```
